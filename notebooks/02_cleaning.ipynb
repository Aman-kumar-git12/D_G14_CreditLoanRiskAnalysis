# Data Cleaning Pipeline

This notebook performs initial cleaning, imputation, and outlier handling for the loan dataset.

## 1. Set Up and Data Loading

```python
import pandas as pd
df = pd.read_csv('../data/raw/loan_data.csv')
df.head()
```

## 2. Basic Cleaning and Row/Column Removal

```python
# # #  Drop rows where 'Loan ID' is null
df.dropna(subset=['Loan ID'], inplace=True)
df.isnull().sum()
```

```python
# Impute numerical columns with median
df['Credit Score'] = df['Credit Score'].fillna(df['Credit Score'].median())
df['Annual Income'] = df['Annual Income'].fillna(df['Annual Income'].median())

```

```python
# Impute categorical and discrete columns with mode
df['Years in current job'] = df['Years in current job'].fillna(df['Years in current job'].mode()[0])
df['Bankruptcies'] = df['Bankruptcies'].fillna(df['Bankruptcies'].mode()[0])
df['Tax Liens'] = df['Tax Liens'].fillna(df['Tax Liens'].mode()[0])
```

```python
# Impute numerical columns with median
df['Credit Score'].fillna(df['Credit Score'].median(), inplace=True)
df['Annual Income'].fillna(df['Annual Income'].median(), inplace=True)
```

```python
# Impute categorical and discrete columns with mode
df['Years in current job'].fillna(df['Years in current job'].mode()[0], inplace=True)
df['Bankruptcies'].fillna(df['Bankruptcies'].mode()[0], inplace=True)
df['Tax Liens'].fillna(df['Tax Liens'].mode()[0], inplace=True)
```

## 4. Handling Outliers and Sentinel Values (Clipping)

```python
# Handle outliers / sentinel values in Current Loan Amount
# Detect 99999999 and clip to 99th percentile of non-sentinel values
loan_upper_limit = df[df['Current Loan Amount'] < 99999999]['Current Loan Amount'].quantile(0.99)
df['Current Loan Amount'] = df['Current Loan Amount'].clip(upper=loan_upper_limit)
print(f'Current Loan Amount clipped at: {loan_upper_limit}')
```

## 6. Cleaning Summary

```python
# Handle outliers in Monthly Debt - Clip to 99th percentile
debt_upper_limit = df['Monthly Debt'].quantile(0.99)
df['Monthly Debt'] = df['Monthly Debt'].clip(upper=debt_upper_limit)
print(f'Monthly Debt clipped at: {debt_upper_limit}')
```

```python
print(f"Final Dataset Shape: {df.shape}")
print(f"Remaining Nulls:\n{df.isnull().sum()}")
```

## 5. Data Export

```python
# --- COMPREHENSIVE DATA CLEANING FIXES ---
# 1. Fill 'Months since last delinquent' to achieve 0 nulls
df['Months since last delinquent'] = df['Months since last delinquent'].fillna(0)

# 2. Fix Credit Score anomalies (> 850)
df.loc[df['Credit Score'] > 850, 'Credit Score'] /= 10

# 3. Standardize Categories
df['Home Ownership'] = df['Home Ownership'].replace('HaveMortgage', 'Home Mortgage')
df['Purpose'] = df['Purpose'].str.lower()

# 4. Convert Years in job to numeric
df['Years in current job'] = df['Years in current job'].astype(str).str.extract('(\d+)').astype(float).fillna(0)

print("Remaining Nulls:")
print(df.isnull().sum())
print("\nUnique Purposes:", df['Purpose'].nunique())
print("Max Credit Score:", df['Credit Score'].max())
```

```python
# Save the refined dataset for EDA and Dashboarding
df.to_csv('../data/processed/cleaned_loan_data.csv', index=False)
```
